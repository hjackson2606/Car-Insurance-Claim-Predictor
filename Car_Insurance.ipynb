## Data Preprocessing

```python
car.info()
```

### Fill NA Values

```python
car.isna().sum()
```

```python
from sklearn.impute import SimpleImputer

num_imputer = SimpleImputer(strategy="median")
car[numerical_cols] = num_imputer.fit_transform(car[numerical_cols])

cat_imputer = SimpleImputer(strategy="most_frequent")
car[categorical_cols] = cat_imputer.fit_transform(car[categorical_cols])
```

```python
car.isna().sum()
```

### Binary Encoding

```python
binary_cols = ["is_esc", "is_adjustable_steering", "is_tpms",
               "is_parking_sensors", "is_parking_camera",
               "is_front_fog_lights", "is_rear_window_wiper",
               "is_rear_window_washer", "is_rear_window_defogger",
               "is_brake_assist", "is_power_door_locks",
               "is_central_locking", "is_power_steering",
               "is_driver_seat_height_adjustable",
               "is_day_night_rear_view_mirror",
               "is_ecw", "is_speed_alert"]

for col in binary_cols:
    car[col] = car[col].map({"Yes":1, "No":0})
```

```python
car.head()
```

### One-Hot Encoding

```python
categorical_cols_wo_binary = [
    "area_cluster", "segment", "model", "fuel_type",
    "max_torque", "max_power", "engine_type",
    "rear_brakes_type", "transmission_type", "steering_type",
]
car = pd.get_dummies(car, columns=categorical_cols_wo_binary, dtype = int, drop_first=True)
```

```python
car.head()
```

```python
car.describe()
```

### Standardize Data

```python
scale_cols = [
    "policy_tenure", "age_of_car", "age_of_policyholder",
    "population_density", "airbags", "displacement",
    "cylinder", "gear_box", "turning_radius",
    "length", "width", "height", "gross_weight", "ncap_rating"
]

from sklearn.preprocessing import StandardScaler
scaler = StandardScaler()
car[scale_cols] = scaler.fit_transform(car[scale_cols])
```

```python
car.describe()
```

## Feature Selection

```python
from sklearn.model_selection import train_test_split
X = car.drop("is_claim", axis=1)
y = car["is_claim"]
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2, random_state=1, stratify=y)
```

```python
print(y_train.value_counts())
```

```python
from imblearn.over_sampling import SMOTE
sm = SMOTE(sampling_strategy="minority")
sm_X_train, sm_y_train = sm.fit_resample(X_train, y_train)
print(sm_y_train.value_counts())
```

## Model Selection, Training, and Testing

### Logistic Regression

```python
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import confusion_matrix, classification_report

LogR = LogisticRegression(max_iter=10000, random_state=1)
LogR.fit(sm_X_train, sm_y_train)
y_pred = LogR.predict(X_test)

print(confusion_matrix(y_test, y_pred))
print(classification_report(y_test, y_pred))
```

### Random Forest

```python
from sklearn.ensemble import RandomForestClassifier
from sklearn.model_selection import GridSearchCV
forest = RandomForestClassifier(random_state=1)
param_grid = {
    "max_features": [32,38,42,43,50],
    "min_samples_leaf": [1,2,3],
    "min_samples_split": [2,3,4],
    "max_depth": [2]
}
grid = GridSearchCV(estimator=forest, param_grid=param_grid, cv=5, scoring="f1", n_jobs=-1, verbose=1)
grid.fit(sm_X_train, sm_y_train)
best_forest = grid.best_estimator_
y_pred = best_forest.predict(X_test)
print("Confusion Matrix:\n", confusion_matrix(y_test, y_pred))
print("Classification Report:\n", classification_report(y_test, y_pred))
print("Best parameters:", grid.best_params_)
```
